```python
# euler_steps.py
import numpy as np
from dynamics import (
    f_first_gear, f_second_gear, f_transition,
    f_aug_first_gear, f_aug_second_gear, f_aug_transition
)

# --------------------------------------
# 0) Rutina genérica (cualquier dinámica vectorial)
# --------------------------------------
def euler_step_vector(f, y, u, dt):
    """
    Paso de Euler explícito genérico.
    Args:
        f  : función f(y, u) → y_dot (np.ndarray o escalar)
        y  : estado en t_k       (np.ndarray o escalar)
        u  : control en t_k      (mismo tipo que espera f)
        dt : Δt                  (float)
    Returns:
        y_next : estado en t_{k+1}
    """
    return y + dt * f(y, u)



# --------------------------------------
# 4)  Dinámicas aumentadas
# --------------------------------------
def euler_step_aug_first_gear(y, T_M, dt):
    """y = [v, z]"""
    return euler_step_vector(f_aug_first_gear, np.asarray(y), T_M, dt)

def euler_step_aug_second_gear(y, T_M, dt):
    """y = [v, z]"""
    return euler_step_vector(f_aug_second_gear, np.asarray(y), T_M, dt)

def euler_step_aug_transition(y, u, dt):
    """
    y = [omega_S, omega_R, z]
    u = [T_M, T_BS, T_BR]
    """
    return euler_step_vector(f_aug_transition, np.asarray(y), np.asarray(u), dt)
```
